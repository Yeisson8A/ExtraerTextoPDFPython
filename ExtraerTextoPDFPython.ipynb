```python
import pytesseract
import shutil
import os
import random
from PIL import Image
import urllib.request
import matplotlib.pyplot as plt
```

```python
# Obtener imagen
urllib.request.urlretrieve('https://metalbyexample.com/wp-content/uploads/figure-65.png', "figure-65.png")
img = Image.open("figure-65.png")
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
# Leer imagen y extraer texto
extractedInformation = pytesseract.image_to_string(img)
print(extractedInformation)
```
